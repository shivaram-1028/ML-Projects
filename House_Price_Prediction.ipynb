```python
X = house_price_dataset.drop(columns=['price'], axis=1)  
Y = house_price_dataset['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
train_prediction = model.predict(X_train)
train_1 = metrics.r2_score(Y_train,train_prediction)
train_2 = metrics.mean_absolute_error(Y_train,train_prediction)
print(train_1,train_2)
```

```python
test_prediction = model.predict(X_test)
test_accuracy = metrics.r2_score(Y_test, test_prediction)
test_accuracy
```
